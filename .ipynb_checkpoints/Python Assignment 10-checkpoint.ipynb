1. Download a picture programmatically
2. Make a numpy-array of all the pixelcolors.
3. Inverse all the colorvalues, so that the image becomes a "negative image"

```python
import invert_colors as ic
import matplotlib.pyplot as plt
import cv2

img_url = "https://images.unsplash.com/photo-1556162748-a6ae062a7fa3"
fn = ic.download_image(img_url)
image = cv2.imread(fn)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.title('Bridge')
plt.imshow(image, interpolation='none')
```

```python
ic.inverte(image, fn)

image = cv2.imread(fn)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title('Inverted Bridge')
plt.imshow(image, interpolation='none')
```
